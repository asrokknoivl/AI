# Filtering

```python
df.loc[(df['y'] == 0) & (df['x1'] < 0.5)] #filtering data with and
```

```python
df.loc[(df['y'] == 1) | (df['x2'] < 0.2)] #filtering with or
```

```python
#resetting indices

new_df = df.loc[(df['y'] == 1) | (df['x2'] < 0.2)] 
new_df.reset_index(drop = True, inplace = True) #dropping the old indices, hence updating. true inplace saves from creating new var
new_df
#otherwise
# new_df = new_df.reset_index(drop = True) #dropping the old indices, hence updating.
# new_df
```

```python
#regex filtering

df = pd.read_csv('Pokemon.csv')
df.loc[df['Name'].str.contains('Mega')]
```

```python
#regex filtering, ~ is similar to NOT in this case

df = pd.read_csv('Pokemon.csv')
df.loc[~df['Name'].str.contains('Mega')]
```

```python
#filtering regex 2
import re
df.loc[df['Type 1'].str.contains('Fire|Grass', regex = True)]
```

```python
#filtering regex 3, re.I means ignore case
df.loc[df['Type 1'].str.contains('fire|grass', flags = re.I, regex = True)]
```

```python
df.loc[df['Name'].str.contains('^pi', flags = re.I, regex = True)]
```

```python
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'FLAAAAAME'
df
```

```python
df.loc[df['Type 1'] == 'WASSS', ['Type 1', 'Type 2']] = ['WASSS','UPUPUPU']
df
```

# Aggregate statistics

```python
#Averaging

df = pd.read_csv('Pokemon.csv')
df.groupby('Type 1').mean().sort_values('Speed', ascending = False) 
```

```python
#Summing

df.groupby('Type 1').sum()
```

```python
#Counting

df.groupby('Type 1').count()
```

```python
#accurate counting 

df['count'] = 1 #can be any value 
df.groupby('Type 1').count()['count']
```

```python
#multitype filtering
df.groupby(['Type 1','Type 2']).count()['count']
```
